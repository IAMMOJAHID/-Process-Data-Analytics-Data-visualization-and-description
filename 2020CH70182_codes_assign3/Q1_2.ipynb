```python
data1 = "data1.xlsx"
read1 = pd.read_excel(data1)
#Result in data1
list_x1 = list(read1['Variable 1'])
#Result2 in data1
list_y1= list(read1['Variable 2'])
# scatter plot for data
fig = plt.figure(figsize = (5, 5))
plt.scatter(list_x1, list_y1, s = 5, c = "blue")  
plt.xlabel("Variable 1")
plt.ylabel("Variable 2")
plt.title("data1")
```

```python
from sklearn.cluster import KMeans
input1=[]
n=len(list_x1)
for i in range(n):
    k=[]
    k.append(list_x1[i])
    k.append(list_y1[i])
    input1.append(k)

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(input1)

gx1=[]
gy1=[]

gx2=[]
gy2=[]

for i in range(n):
    if(kmeans.labels_[i]==0):
        gx1.append(list_x1[i])
        gy1.append(list_y1[i])
    else:
        gx2.append(list_x1[i])
        gy2.append(list_y1[i])

```

```python
centroids = kmeans.cluster_centers_
labels = kmeans.predict(X)
```

```python
plt.scatter(list_x1, list_y1, c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=100, linewidths=1, color='red')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('KMeans Clustering with k=2')
plt.show()
```
