```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

%matplotlib inline
from scipy.stats import kurtosis
from scipy.stats import skew
from statsmodels import robust
import statistics 
import time
```

```python
start_time = time.time()
data1 = "data.xlsx"
read1 = pd.read_excel(data1)
#Result in data1
list_x1 = list(read1['Result 1'])
#Result2 in data1
list_y1= list(read1['Result 2'])
# scatter plot for data
fig = plt.figure(figsize = (5, 5))
plt.scatter(list_x1, list_y1, s = 5, c = "blue")  
plt.xlabel("Result 1")
plt.ylabel("Result 2")
plt.title("data")
plt.show()
```

```python
# b
from sklearn.cluster import KMeans
input1=[]
n=len(list_x1)
for i in range(n):
    k=[]
    k.append(list_x1[i])
    k.append(list_y1[i])
    input1.append(k)

kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(input1)

gx1=[]
gy1=[]

gx2=[]
gy2=[]

for i in range(n):
    if(kmeans.labels_[i]==0):
        gx1.append(list_x1[i])
        gy1.append(list_y1[i])
    else:
        gx2.append(list_x1[i])
        gy2.append(list_y1[i])

fig = plt.figure(figsize = (7, 7))
plt.scatter(gx1, gy1, s = 5, c = "green")
plt.scatter(gx2, gy2, s = 5, c = "blue")
plt.xlabel("Result 1")
plt.ylabel("Result 2")
plt.title("K-Means clustering")
plt.show()
run_time1=(time.time() - start_time)
```

```python
# c
start_time2 = time.time()
data2 = "data_GMM.xlsx"
read2 = pd.read_excel(data2)
#Result in data1
list_x2 = list(read2['Result 1'])
#Result2 in data1
list_y2= list(read2['Result 2'])

# scatter plot for data_GMM
fig = plt.figure(figsize = (5, 5))
plt.scatter(list_x2, list_y2, s = 5, c = "green")  
plt.xlabel("Result 1")
plt.ylabel("Result 2")
plt.title("data_GMM")
plt.show()
```

```python
# d
from sklearn.mixture import GaussianMixture
input_GM=[]
n=len(list_x1)
for i in range(n):
    k=[]
    k.append(list_x2[i])
    k.append(list_y2[i])
    input_GM.append(k)
    
gmm = GaussianMixture(n_components=2)
gmm.fit(input_GM)
labels = gmm.predict(input_GM)

Gx1=[]
Gy1=[]

Gx2=[]
Gy2=[]

for i in range(n):
    if(labels[i]==0):
        Gx1.append(list_x2[i])
        Gy1.append(list_y2[i])
    else:
        Gx2.append(list_x2[i])
        Gy2.append(list_y2[i])
        
fig = plt.figure(figsize = (7, 7))
plt.scatter(Gx1, Gy1, s = 5, c = "green")
plt.scatter(Gx2, Gy2, s = 5, c = "blue")
plt.xlabel("Result 1")
plt.ylabel("Result 2")
plt.title("Gaussian Mixture Model")
plt.show()
run_time2=(time.time() - start_time2)
# print("%s seconds" % (time.time() - start_time2))
```

```python
# d
print("Run time of first method: %sseconds" % (run_time1))
print("Run time of second method: %sseconds" % (run_time2))
```
