```python
# Outliers for data1 using MAD
```

```python
def median(x):
    
    xs = sorted(x)
    
    n = len(xs)
    
    if (n%2)!=0:
        md = xs[(n + 1)/2 - 1]
    else:
        md = (xs[int(n/2)] + xs[int(n/2) - 1])/2
    return md
```

```python
median_x1=read1['x'].median()
median_y1=read1['y'].median()
temp_x1=[]
temp_y1=[]
for i in range(len(list_x1)):
    temp_x1.append(abs(list_x1[i]-median_x1))
    temp_y1.append(abs(list_y1[i]-median_y1))
MAD_x1=median(temp_x1)
MAD_y1=median(temp_y1)
```

```python
nrml_x1=[]
ot_x1=[]
nrml_y1=[]
ot_y1=[]
for i in range(len(list_x1)):
    if( abs(0.6745*(list_x1[i]-median_x1)/MAD_x1)>3 or abs(0.6745*(list_y1[i]-median_y1)/MAD_y1)>3 ):
        ot_x1.append(list_x1[i])
        ot_y1.append(list_y1[i])
    else:
        nrml_x1.append(list_x1[i])
        nrml_y1.append(list_y1[i])
```

```python
fig = plt.figure(figsize = (10, 10))
plt.scatter(ot_x1, ot_y1, s = 8, c = "red")  
plt.scatter(nrml_x1, nrml_y1, s = 3, c = "blue")
plt.show()
```
